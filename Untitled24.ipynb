**Q1. Difference between descriptive and inferential statistics.**

**Ans:** Descriptive statistics summarize data using measures like mean,median,and graphs(e.g.,average marks of one class).

Inferential statistics use sample data to make conclusions or predictions about a population (e.g.,predicting all students'average from one class).



**Q2. What is sampling? Difference between random and stratified sampling.**

**Ans:** Sampling means selecting a part of the popultion to study instead of the whole.



*   Random Sampling:Every individual has an equal chance instead of the whole.
*   Stratified Sampling: Population is divided into groups(strata)and samples are taken from each group.

**Example:** Random-choosing any 50 students;Stratified choosing students from each grade.

**Q4. Explain skewnees and kurtosis. What does a positive skew imply about the data?**

**Ans:**

**Q5. Implement a Python program to compute the mean, median and mode of a given list of numbers.**

**Ans: **

```python
import statistics as stats

numbers=[12,15,12,18,19,12,20,22,19,19,24,24,24,26,28]
mean=stats.mean(numbers)
median=stats.median (numbers)
mode=stats.mode(numbers)

print("Mean:",mean)
print("Median:",median)
print("Mode(s):",mode)
```

**Q6.Compute the covariance and correlation coefficient between the two datasets.**

**Ans:**

```python
import numpy as np

list_x=[10,20,30,40,50]
list_y=[15,25,35,45,60]

covariance=np.cov(list_x,list_y,bias=False)[0][1]

correlation=np.corrcoef(list_x,list_y)[0][1]

print("Covariance:",covariance)
print("Correlation Coefficient:",correlation)
```

**Explanation:**
Covariance shows that both variables increase together.Correlation (close to1)means they have a strong positive relationship.

**Q8. Relationship between advertising spend and daily sales.**

**Ans: Explanation:** Covariance and correlation help to check  how two variables move together.

* Covariance tells the direction of the relationship(positive or negative).
* Correlation tells the strength of that relationship(value between -1 and +1).

If correlation is close to +1,it means higher ad spend increases sales.



```python
import numpy as np

# Data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print(f"Correlation between advertising spend and daily sales: {correlation:.2f}")
```

**Q9. Analyze customer satisfaction survery data.**

**Ans:**  **Explanation:** We can understand the data using:

*  Mean & Standard Deviation-show average satisfaction and variation.
*  Histogram- shows how scores are distributed (which range is most common).



```python
import matplotlib.pyplot as plt
import statistics as stats

survey_scores=[7,8,5,9,6,7,8,9,10,4,7,6,9,8,7]

mean=stats.mean(survey_scores)
std_dev=stats.stdev(survey_scores)

print("Mean:",mean)
print("Standard Deviation:",std_dev)

plt.hist(survey_scores,bins=5,edgecolor='black')
plt.title("Customer Satisfaction Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
